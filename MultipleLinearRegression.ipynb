Multiple linear regression Impementation using python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
dataset = pd.read_csv('Advertising.csv')
x = dataset[['TV', 'radio', 'newspaper']]
y = dataset['sales']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```python
print("Intercept:", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

```python
# prediction of the test set
y_pred_mlr = mlr.predict(x_test)

# predicted values
print("Predicted values:", y_pred_mlr)
```

```python
mlr_diff = pd.DataFrame({'Actual values': y_test, 'Predicted values': y_pred_mlr})
mlr_diff.head()
```

```python
import matplotlib.pyplot as plt

# Assuming y_test and y_pred are 1D
plt.scatter(y_test, y_pred_mlr)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'red')
plt.show()
```

```python
# model evaluation
from sklearn import metrics

meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))

print("R squared: {:.2f}".format(mlr.score(x, y)*100))
print("Mean Absolute Error:", meanAbErr)
print("Mean Squared Error:", meanSqErr)
print("Root Mean Squared Error:", rootMeanSqErr)
```
